```python
q1= d['FE'].quantile(0.25)
q2= d['FE'].quantile(0.50)
q3= d['FE'].quantile(0.75)
IQR= q3-q1
ul= q3+(1.5*IQR)
ll= q3-(1.5*IQR)
print(np.where(d['FE']>ul))
```
